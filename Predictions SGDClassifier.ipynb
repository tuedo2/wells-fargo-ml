```python
# import dataset
import pandas as pd
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# Ignore useless warnings (see SciPy issue #5998)
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")

df = pd.read_excel(r'/home/tuedo/Downloads/ChallengeDataSet-WellsFargo.xlsx') # path to training set
df1 = df.drop('XC', axis=1) # no XC column at first
X = df1.to_numpy()
df2 = df["y"]
y = df2.to_numpy()
# we use numpy arrays instead of dataframes because I'm more comfortable with them

# visualization of training dataset
print(df1)
print(df2)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.base import clone

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X, y) # initially fit all datapoints, and then we refine within specific classes (A, B, C, D, E)
```

```python
ind_A = np.where(df["XC"]=='A')
train_A = X[np.ma.masked_array(ind_A[0]), :]
X_A, y_A = train_A[:, 0:30], train_A[:, 30]
clf_A = clone(sgd_clf)
clf_A.fit(X_A, y_A)

ind_B = np.where(df["XC"]=='B')
train_B = X[np.ma.masked_array(ind_B[0]), :]
X_B, y_B = train_B[:, 0:30], train_B[:, 30]
clf_B = clone(sgd_clf)
clf_B.fit(X_B, y_B)

ind_C = np.where(df["XC"]=='C')
train_C = X[np.ma.masked_array(ind_C[0]), :]
X_C, y_C = train_C[:, 0:30], train_C[:, 30]
clf_C = clone(sgd_clf)
clf_C.fit(X_C, y_C)

ind_D = np.where(df["XC"]=='D')
train_D = X[np.ma.masked_array(ind_D[0]), :]
X_D, y_D = train_D[:, 0:30], train_D[:, 30]
clf_D = clone(sgd_clf)
clf_D.fit(X_D, y_D)

ind_E = np.where(df["XC"]=='E')
train_E = X[np.ma.masked_array(ind_E[0]), :]
X_E, y_E = train_E[:, 0:30], train_E[:, 30]
clf_E = clone(sgd_clf)
clf_E.fit(X_E, y_E)
```

```python
eval_df = pd.read_excel(r'/home/tuedo/Downloads/EvaluationDataSet-WellsFargo.xlsx') # path to eval set
eval_df
```

```python
data_eval = eval_df.drop("XC", axis=1).to_numpy()
data_eval.shape
```

```python
X_eval = data_eval[:, 0:30]
```

```python
# predicting subsets of the test set based on 'XC' column
eval_ind_A = np.where(eval_df["XC"]=='A')
eval_ind_B = np.where(eval_df["XC"]=='B')
eval_ind_C = np.where(eval_df["XC"]=='C')
eval_ind_D = np.where(eval_df["XC"]=='D')
eval_ind_E = np.where(eval_df["XC"]=='E')

eval_A = X_eval[np.ma.masked_array(eval_ind_A[0]), :]
eval_B = X_eval[np.ma.masked_array(eval_ind_B[0]), :]
eval_C = X_eval[np.ma.masked_array(eval_ind_C[0]), :]
eval_D = X_eval[np.ma.masked_array(eval_ind_D[0]), :]
eval_E = X_eval[np.ma.masked_array(eval_ind_E[0]), :]
```

```python
predict_A = clf_A.predict(eval_A)
predict_B = clf_B.predict(eval_B)
predict_C = clf_C.predict(eval_C)
predict_D = clf_D.predict(eval_D)
predict_E = clf_E.predict(eval_E)
predictions = np.zeros(7000)
predictions
```

```python
# merging all of our predictions into one array again
ax, bx, cx, dx, ex = 0, 0, 0, 0, 0
i = 0;
for xc in eval_df["XC"]:
    if (xc == "A"):
        predictions[i] = predict_A[ax]
        ax += 1
    elif (xc == "B"):
        predictions[i] = predict_B[bx]
        bx += 1
    elif (xc == "C"):
        predictions[i] = predict_C[cx]
        cx += 1
    elif (xc == "D"):
        predictions[i] = predict_D[dx]
        dx += 1
    elif (xc == "E"):
        predictions[i] = predict_E[ex]
        ex += 1
        
    i += 1

predictions.T
```

```python
predictions
```

```python
# recasting into DataFrame

pred_df = pd.DataFrame(data=predictions, columns=["prediction_score"])
pred_df.insert(0, "dataset_id", pd.Series(range(1,7001)), allow_duplicates=False)
```
